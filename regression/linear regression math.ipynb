This is a simple linear regression, to learn about how linear regression works with math stand point.

the main function here is $$Y = wX + b$$

whereas Y is the output, W is the weight(slope), B is the bias and X is the input

we need to find the best W and B, which means we need to find a line which minimizes error.

error = Y actual - Y predicted

We don’t minimize raw error.
We minimize squared error.

$$Cost\;Function = J(w, b) = \sum_{i=1}^{n} (y^{(i)} - (wx^{(i)} + b))^2$$


$$Y = wX + b$$

$$
\underbrace{\begin{bmatrix} \hat{y}_1 \\ \hat{y}_2 \\ \vdots \\ \hat{y}_m \end{bmatrix}}_{\mathbf{\hat{y}}} =
\underbrace{\begin{bmatrix} x_1 & 1 \\ x_2 & 1 \\ \vdots & \vdots \\ x_m & 1 \end{bmatrix}}_{\mathbf{X}} \times
\underbrace{\begin{bmatrix} w \\ b \end{bmatrix}}_{\mathbf{\theta}}
$$

(Note: Here, the bias $b$ is often folded into the weight vector by adding a column of 1s to the input matrix $X$.)

while multiplication, it will be: $$\hat{y}_1 = w*x_1 + 1*b$$

$$Analytical\;formula : \mathbf{w} = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T\mathbf{y}$$

## 1. Defining the Constants

In the matrix world, we represent our data as follows:

- $\mathbf{y}$: The Actual Output vector ($m \times 1$). Contains the true labels.
- $\mathbf{X}$: The Input Matrix ($m \times n$). Each row is a training example; each column is a feature.
- $\theta$: The Parameter Vector (Weights). This contains both $w$ and $b$.
- $\hat{y}$: The Predicted Output. Calculated as $\hat{y} = \mathbf{X}\theta$.
- $\mathbf{e}$: The Error Vector $(\mathbf{y} - \hat{y})$. The difference between reality and prediction for every data point.

## 2. From Scalar to Matrix: The Sum of Squares

Why does $\sum(y - \hat{y})^2$ become $\mathbf{e}^T\mathbf{e}$?

In scalar math, to sum the squares of a list of numbers, you square each one and add them. In linear algebra, the Dot Product of a vector with itself accomplishes the exact same thing.

If $\mathbf{e} = \begin{bmatrix} e_1 \\ e_2 \end{bmatrix}$, then:

$$\mathbf{e}^T\mathbf{e} = \begin{bmatrix} e_1 & e_2 \end{bmatrix} \begin{bmatrix} e_1 \\ e_2 \end{bmatrix} = (e_1 \cdot e_1) + (e_2 \cdot e_2) = \sum e_i^2$$

This is why we define the Cost Function $J(\theta)$ as:

$$J(\theta) = (\mathbf{y} - \mathbf{X}\theta)^T (\mathbf{y} - \mathbf{X}\theta)$$

## 3. The Derivation of the Normal Equation

To find the $\theta$ that minimizes the error, we expand the cost function and use calculus.

### Step 1: Expand the Product

Using the property $(\mathbf{A}-\mathbf{B})^T = \mathbf{A}^T - \mathbf{B}^T$:

$$J(\theta) = (\mathbf{y}^T - \theta^T \mathbf{X}^T)(\mathbf{y} - \mathbf{X}\theta)$$

$$J(\theta) = \mathbf{y}^T\mathbf{y} - \mathbf{y}^T\mathbf{X}\theta - \theta^T\mathbf{X}^T\mathbf{y} + \theta^T\mathbf{X}^T\mathbf{X}\theta$$

Note: Since the result of each term is a scalar, $\mathbf{y}^T\mathbf{X}\theta$ is equal to its transpose $\theta^T\mathbf{X}^T\mathbf{y}$. We can combine them:

$$J(\theta) = \mathbf{y}^T\mathbf{y} - 2\theta^T\mathbf{X}^T\mathbf{y} + \theta^T\mathbf{X}^T\mathbf{X}\theta$$

### Step 2: Take the Derivative

We take the partial derivative with respect to $\theta$:

$$\frac{\partial J}{\partial \theta}$$

- The derivative of $\mathbf{y}^T\mathbf{y}$ is $0$ (it doesn't depend on $\theta$).
- The derivative of $2\theta^T\mathbf{X}^T\mathbf{y}$ is $2\mathbf{X}^T\mathbf{y}$.
- The derivative of the quadratic term $\theta^T\mathbf{X}^T\mathbf{X}\theta$ is $2\mathbf{X}^T\mathbf{X}\theta$.

Set the derivative to 0 to find the minimum:

$$-2\mathbf{X}^T\mathbf{y} + 2\mathbf{X}^T\mathbf{X}\theta = 0$$

### Step 3: Solve for $\theta$

Move the terms:

$$\mathbf{X}^T\mathbf{X}\theta = \mathbf{X}^T\mathbf{y}$$

Multiply both sides by the inverse $(\mathbf{X}^T\mathbf{X})^{-1}$ to isolate $\theta$:

$$\theta = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$$

In our expansion, we have two middle terms: $\mathbf{y}^T\mathbf{X}\theta$ and $\theta^T\mathbf{X}^T\mathbf{y}$.

## 1. The Dimensional Proof

Let's check the "shape" of these terms to prove they are just single numbers (scalars).

Assume we have $m$ examples and $n$ features:

- $\mathbf{y}^T$ is $(1 \times m)$
- $\mathbf{X}$ is $(m \times n)$
- $\theta$ is $(n \times 1)$

Multiplying them: $(1 \times m) \times (m \times n) \times (n \times 1) = \mathbf{(1 \times 1)}$.

Since the result is a scalar, we can say:

$$\text{Scalar} = (\text{Scalar})^T$$

## 2. The Transpose Rule

There is a property in linear algebra: $(\mathbf{ABC})^T = \mathbf{C}^T\mathbf{B}^T\mathbf{A}^T$.

If we take the transpose of the first middle term:

$$(\mathbf{y}^T\mathbf{X}\theta)^T = \theta^T\mathbf{X}^T(\mathbf{y}^T)^T = \theta^T\mathbf{X}^T\mathbf{y}$$

Because the transpose of the first term is the second term, and both are just single numbers, they are identical.

$$A + A = 2A$$

That is how we get $-2\theta^T\mathbf{X}^T\mathbf{y}$.

## 3. Numerical Example

Let's use tiny matrices to see it in action.

Let $\mathbf{y} = \begin{bmatrix} 2 \end{bmatrix}$, $\mathbf{X} = \begin{bmatrix} 3 \end{bmatrix}$, and $\theta = \begin{bmatrix} 4 \end{bmatrix}$.

Calculate the first term $\mathbf{y}^T\mathbf{X}\theta$:

$$\begin{bmatrix} 2 \end{bmatrix} \times \begin{bmatrix} 3 \end{bmatrix} \times \begin{bmatrix} 4 \end{bmatrix} = 2 \times 3 \times 4 = \mathbf{24}$$

Calculate the second term $\theta^T\mathbf{X}^T\mathbf{y}$:

$$\begin{bmatrix} 4 \end{bmatrix} \times \begin{bmatrix} 3 \end{bmatrix} \times \begin{bmatrix} 2 \end{bmatrix} = 4 \times 3 \times 2 = \mathbf{24}$$

Even if the matrices were huge, the result of that specific multiplication string always collapses into one single number representing a part of the "total error." Since they both equal 24, we just write $2 \times 24$.

```python
import numpy as np
from matplotlib import pyplot as plt
np.random.seed(42)
```

```python
X = 2 * np.random.rand(100, 1)
# X
```

```python
y = 4 + 3 * X + np.random.randn(100, 1)
# y
```

```python
plt.scatter(x=X, y=y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('Visualisation of data')
```

```python
X_b = np.c_[X, np.ones((100, 1))]
```

Since we are using the matrix multiplication with numpy, we need to add '1' to every row, to make the matrix multiplication possible.

example
[[1.43, 1],
 [0.23, 1], 
 ...]

```python
# Xt X inverse
Xt = X_b.T
# print(Xt)
Xt_X_inv = np.linalg.inv(Xt.dot(X_b))
# print(Xt_X_inv)
theta = Xt_X_inv.dot(Xt).dot(y)

print("W = ", theta[0])
print("B = ", theta[1])
```

Why Does This Formula Work? Because we minimized: $$\sum (y - X\theta)^2$$Taking derivative w.r.t $\theta$:$$X^T X \theta = X^T y$$Solve for $\theta$:$$\theta = (X^T X)^{-1} X^T y$$That’s the origin.

```python
y_pred = X_b.dot(theta)
plt.scatter(x=X, y=y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('Visualisation of data')
plt.plot(X, y_pred, color='red')
```
