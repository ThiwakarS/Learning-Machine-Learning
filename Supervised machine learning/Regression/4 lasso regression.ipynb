# Regularization in Linear Regression

When a model is too complex or features are correlated, ordinary least squares tends to overfit — it chases noise in the training data and produces unreliable weights. Regularization fixes this by adding a **penalty term** to the cost function that discourages large weights. The three main approaches are Ridge, Lasso, and Elastic Net.

## 1. Lasso Regression (L1 Regularization)

### The Cost Function

Lasso (Least Absolute Shrinkage and Selection Operator) penalizes the **absolute value** of the weights instead of the square:

$$J(\theta) = \|\mathbf{y} - \mathbf{X}\theta\|^2 + \lambda\sum|\theta_i|$$

Which is written compactly as:

$$J(\theta) = \|\mathbf{y} - \mathbf{X}\theta\|^2 + \lambda\|\theta\|_1$$

The subscript $1$ denotes the **L1 norm** — the sum of absolute values.

### Why Is This Different From Ridge?

The L1 penalty causes **sparsity**. Some weights are pushed to be **exactly zero**, which means those features are completely ignored by the model. This is automatic feature selection — Lasso figures out which features matter and which do not.

Ridge shrinks weights but keeps all of them. Lasso shrinks some weights all the way to zero and keeps the rest.

### No Closed Form Solution

Unlike Ridge, Lasso has **no analytical solution**. This is because the absolute value function $|x|$ is not differentiable at $x = 0$ — you cannot take a clean derivative and set it to zero. Instead, Lasso is solved numerically using methods like coordinate descent.

### Geometric Intuition

Lasso is equivalent to minimizing OLS subject to:

$$\|\theta\|_1 \leq c$$

In 2D, this constraint is a **diamond** (rotated square). The corners of the diamond lie exactly on the axes, meaning one weight is zero and the other is not. When the OLS error surface meets the constraint region, it tends to hit a **corner** rather than a smooth edge — and at a corner, one or more weights are exactly zero.

This is the geometric reason Lasso produces sparse solutions while Ridge does not.

### When To Use Lasso

**Use Lasso when:**
- You suspect many features are irrelevant
- You want the model to automatically select important features
- Working with high-dimensional data where most features are noise

**Avoid Lasso when:**
- Features are highly correlated — Lasso will arbitrarily pick one and zero out the others, which can be unstable

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

np.random.seed(42)

# 100 samples, 3 features
X = np.random.rand(100, 3)

# True relationship:
# y = 5*x1 + 3*x2 - 2*x3 + 4
true_weights = np.array([5, 3, -2])
y = X @ true_weights + 4

# X @ true_weights, here it represents the matrix multiplication
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

lasso = make_pipeline(
    StandardScaler(),
    Lasso(alpha=0.1)
)
lasso.fit(X_train, y_train)
```

```python
y_pred = lasso.predict(X_test)
print("R2:", r2_score(y_test, y_pred))
print("Weights:", lasso.named_steps["lasso"].coef_)
print("intercept:", lasso.named_steps["lasso"].intercept_)
```
